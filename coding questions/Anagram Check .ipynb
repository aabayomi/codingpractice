```python
def anagram_check(S1,S2):
    
    S1 = S1.replace(' ','').lower()  ## special case to ignore spacing 
    S2 = S2.replace(' ','').lower()
    # Edge case Check
    if len(S1)!=len(S2):
        return False
    
    count ={}
    
    for letter in S1:             # O(n) time complexity
        if letter in count:
            count[letter] += 1
        else:
            count[letter] =1
    for letter in S2:
        if letter in count:
            count[letter] -=1
        else:
            count[letter] = 1
    for k in count:
        if count[k] != 0:
            return False
    return True
            
```

```python
anagram_check('dogfri','good')
```

```python
anagram_check('clint eastwood','old west action')
```
