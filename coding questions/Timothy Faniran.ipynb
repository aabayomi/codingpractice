```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plot
import seaborn as sns
import dateutil

import plotly.plotly as py
import plotly.graph_objs as go

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

init_notebook_mode(connected=True)


# plot([go.Scatter(x=[1, 2, 3], y=[3, 1, 6])])


# print plotly.__version__

perfdata=pd.read_csv("/Users/abayomi/Downloads/data.csv")
# data = pd.DataFrame.from_csv('phone_data.csv')

# print perfdata.head()

date_count = perfdata[['Intent_Date','Intent_Completed_Count']]

region_count = perfdata[['Merchant_Region_Name','Intent_Completed_Count']]

# print region_count.head()
# print date_count.head()

print 

# print perfdata.groupby(['Merchant_Region_Name'])['Intent_Completed_Count'].count()

print 

groupbyRegion = perfdata['Intent_Completed_Count'].groupby(perfdata['Merchant_Region_Name'])

groupbyRegion2 =  perfdata.groupby('Merchant_Region_Name').describe()

regions = []

counts_data = []
counts_mean = []

for region, new_df in groupbyRegion2.groupby(level=0):
    regions.append(region)
    
    for col in new_df.columns:
        new_col = new_df[col][region]
        x_label = region
        
#     print new_col
    
        # using the Box method of Plotly's graph objects
        box = go.Box(y= new_col, name=x_label)
#         box = go.Box(y= new_col['min'], name=x_label)
    
        counts_data.append(box)
        counts_mean.append(new_col)

```

```python
counts_data.append(go.Scatter(x=regions, y=counts_mean, mode='lines', name='mean'))


# print counts_data

iplot({
        'data': counts_data,
        'layout': {
            'title': 'Count Comparison',
            'xaxis': {'title': 'Regions'},
            'yaxis': {'title': 'Intent Completed Count'}
        }
})
```
